```python
# Titanic submission using Logistic Regression 
# Kaggle score: 0.74641

import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

def clean_data(df):
    # Drop columns with missing data
    df = df.drop(['Name','Cabin','Ticket','Embarked','Parch'], axis = 1)
    
    # Replace categorical sex data with binary
    df['Sex'] = df['Sex'].replace(['male','female'],[0,1])
    
    # Fill missing age with mean
    df['Age'] = df['Age'].fillna(df['Age'].mean())
    df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
    return df

train_data = clean_data(train_data)
test_data = clean_data(test_data)
```

```python
# Split into data/target data frames
y_train = train_data.pop('Survived')
```

```python
# Creating sklearn model
Log_Reg = LogisticRegression()
Log_Reg = Log_Reg.fit(train_data, y_train)

new_pred = Log_Reg.predict(test_data)
pd.DataFrame({'PassengerId': test_data.PassengerId, 
              'Survived':new_pred}
            ).set_index('PassengerId').to_csv('LR_sub.csv')

# Can now submit LR_sub.csv to kaggle
```
